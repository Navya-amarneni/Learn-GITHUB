```python
import random as rnd
import networkx as nx

G3 = nx.DiGraph()
G3.add_nodes_from(range(15))
probabilty_max=0.6
for i in range(15):
  for j in range(15):
    if(rnd.random()<=probabilty_max):
      G3.add_edge(i,j)

nx.draw(G3,with_labels="true")
```
